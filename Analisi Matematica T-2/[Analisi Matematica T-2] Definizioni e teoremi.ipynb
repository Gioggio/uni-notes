# Spazi metrici e normati

**Definizione** (metrica in $\mathbb{R}^n$) Un'applicazione $d:\mathbb{R}^n\times\mathbb{R}^n\rightarrow\mathbb{R}$ è una **metrica** se soddisfa:

1. $\forall\overline{x} = (x_1,\dots, x_n), \overline{y}=(y_1,\dots, y_n)\in\mathbb{R}^n$ si ha
$$d(\overline{x}, \overline{y}) \geq 0\ \text{e}\ d(\overline{x}, \overline{y}) = 0 \iff \overline{x} = \overline{y}$$

2. $\forall\overline{x},\overline{y}\in\mathbb{R}^n$ si ha
$$d(\overline{x}, \overline{y}) = d(\overline{x}, \overline{y})$$

3. (disuguaglianza triangolare) $\forall\overline{x}, \overline{y}, \overline{z}\in\mathbb{R}^n$ si ha
$$ d(\overline{x}, \overline{y}) \leq d(\overline{x}, \overline{z}) + d(\overline{y}, \overline{z})$$

**Dimostrazione** (disuguaglianza triangolare)

$$\sqrt{(x_1-y_1)^2 + (x_2-y_2)^2} \leq \sqrt{(x_1-z_1)^2 + (x_2-z_2)^2} + \sqrt{(y_1-z_1)^2 + (y_2-z_2)^2}$$

Prendiamo $a_i = (x_i-z_i)$ e $b_i = (z_i-y_i)$

$$\sqrt{(a_1+b_1)^2 + (a_2+b_2)^2} \leq \sqrt{a_1^2+a_2^2} + \sqrt{b_1^2+b_2^2} \iff$$

$$\iff (a_1+b_1)^2 + (a_2+b_2)^2 \leq a_1^2 + a_2^2 + b_1^2 + b_2^2 + 2\sqrt{(a_1^2+a_2^2)(b_1^2+b_2^2)} \iff$$

$$\iff a_1b_1+a_2b_2 \leq 2\sqrt{(a_1^2+a_2^2)(b_1^2+b_2^2)}$$

Se $a_1b_1+a_2b_2 \leq 0$ la disuguaglianza è verificata
Se $a_1b_1+a_2b_2 \geq 0$:

$$(a_1b_1+a_2b_2)^2 \leq (a_1^2+a_2^2)(b_1^2+b_2^2) \iff$$

$$\iff 2a_1a_2b_1b_2 \leq a_1^2b_2^2+a_2^2b_1^2 \iff$$

$$\iff (a_1b_2 - a_2b_1)^2 \geq 0$$

**Definizione** (spazio metrico) Sia $X$ un insieme e $d$ una metrica per $X$. Allora $(X, d)$ si dice **spazio metrico**.

**Definizione** (norma di un vettore) Sia $\overline{x} = (x_1,\dots ,x_n)$. Si definisce norma di $\overline{x}$ una funzione $||\cdot||: \mathbb{R}^n\rightarrow\mathbb{R}$ che soddisfa le seguenti proprietà:

1. $||\overline{x}|| \geq 0 \forall\overline{x}\in\mathbb{R}^n$ e $||\overline{x}||=0 \iff \overline{x}=\overline{0}$

2. $\forall\overline{x},\overline{y}\in\mathbb{R}^n\ ||\overline{x}+\overline{y}|| \leq ||\overline{x}|| + ||\overline{y}||$

3. $\forall\lambda\in\mathbb{R},\ \forall\overline{x}\in\mathbb{R}^n\ ||\lambda\overline{x}|| = |\lambda|\cdot||\overline{x}||$

4. $\forall\overline{x}\overline{y}\in\mathbb{R}^n\ d(\overline{x}, \overline{y}) = ||\overline{x} - \overline{y}||$ (con metrica euclidea $d$)

**Definizione** (prodotto scalare) Si definisce **prodotto scalare** una funzione $\langle,\rangle: \mathbb{R}^n \times \mathbb{R}^n\rightarrow\mathbb{R}$ tale che

$$\forall\overline{x}, \overline{y}\in\mathbb{R}^n\ \langle\overline{x},\overline{y}\rangle = \sum_{i=1}^{n}{x_iy_i}$$

e rispetti le seguenti proprietà:

1. $\overline{x}\perp\overline{y} se \langle\overline{x},\overline{y}\rangle = 0$

2. $|\langle\overline{x},\overline{y}\rangle| \leq ||\overline{x}||\cdot||\overline{y}||$

3. $\langle\overline{x},\overline{y}\rangle = ||\overline{x}||\cdot||\overline{y}||\cdot\cos{\theta}$

4. $\langle\overline{x},\overline{x}\rangle = ||\overline{x}||^2$

## Insiemistica

**Definizione** (palla aperta di centro $\overline{x_0}$ e raggio $\delta > 0$)

$$ B_{\delta}(\overline{x_0}) = \{\overline{x}\in\mathbb{R}\ |\ d(\overline{x},\overline{x_0}) < \delta\}$$

**Definizione** (insieme aperto) $A\subseteq\mathbb{R}^n$ è un **insieme aperto** se

$$\forall\overline{x}\in\mathbb{R}^n\ \exists\delta=\delta(\overline{x_0})>0\ \text{e}\ B_{\delta}(\overline{x_0})\subseteq A$$

**Proprietà** (insiemi aperti)

1. $\emptyset$ e $\mathbb{R}^n$ sono insiemi aperti.

2. L'unione di un numero arbitrario di insiemi aperti è ancora un insieme aperto.

$$A = \bigcup_{\alpha}{A_{\alpha}}\ \text{con}\ A_{\alpha}\ \text{insieme aperto e}\ \alpha\in\mathbb{N}\cup\{+\infty\}$$

3. L'intersezione di un numero finito di insiemi aperti è ancora un insieme aperto.

$$A = \bigcap_{\alpha}{A_{\alpha}}\ \text{con}\ A_{\alpha}\ \text{insieme chiuso e}\ \alpha\in\mathbb{N}$$

**Osservazione** $B_{r}(\overline{x_0})$ è un insieme aperto.

**Definizione** (insieme chiuso) $A\subseteq\mathbb{R}^n$ è un **insieme chiuso** se $\mathbb{R}^n \setminus A$ è un insieme aperto.

**Proprietà** (insiemi chiusi)

1. $\emptyset$ e $\mathbb{R}^n$ sono insiemi chiusi.

2. L'unione di un numero arbitrario di insiemi chiusi è ancora un insieme chiuso.

$$A = \bigcap_{\alpha}{A_{\alpha}}\ \text{con}\ A_{\alpha}\ \text{insieme aperto e}\ \alpha\in\mathbb{N}\cup\{+\infty\}$$

3. L'intersezione di un numero finito di insiemi chiusi è ancora un insieme chiuso.

$$A = \bigcup_{\alpha}{A_{\alpha}}\ \text{con}\ A_{\alpha}\ \text{insieme chiuso e}\ \alpha\in\mathbb{N}$$

**Osservazione** Insiemi chiusi in dimensioni inferiori rimangono chiusi in dimensioni superiori.

**Definizione** (punto interno di un insieme) Sia $A\subseteq\mathbb{R}^n$ e $\overline{x_0}\in A$. Chiamiamo $\overline{x_0}$ **punto interno** di A se

$$\exists\delta > 0\ \text{tale che}\ B_{\delta}(\overline{x_0})\subseteq A$$

**Definizione** (interno di un insieme) Sia $A\subseteq\mathbb{R}^n$. Chiamiamo **interno** di $A$ l'insieme dei punti interni di $A$, denotato con $\overset{\circ}{A}$.

**Osservazione** $\overset{\circ}{A}$ è il più grande sottoinsieme di $A$ aperto. Se $A$ non è aperto può succedere che $\overset{\circ}{A} = \emptyset$ (es. se $A=\mathbb{N}$).

**Definizione** (punto aderente di un insieme) Sia $A\subseteq\mathbb{R}^n$. Il punto $\overline{x_0}\in\mathbb{R}^n$ si dice **punto aderente** di $A$ se

$$\forall\delta > 0\ A \cap B_{\delta}(\overline{x_0})\neq\emptyset$$

**Osservazione** Se $\overline{x_0}$ appartiene all'insieme, allora è un punto aderente dell'insieme.

**Definizione** (chiusura di un insieme) Sia $A\subseteq\mathbb{R}^n$. Chiamiamo **chiusura** di $A$ l'insieme dei punti aderenti di $A$, denotato con $\overline{A}$.

1. $A\subseteq\overline{A}$

2. $\overline{A}$ è chiuso.

3. $\mathbb{R}^n \setminus A$ è aperto.

4. Se $A$ è chiuso, allora $A = \overline{A}$.

**Definizione** (punto di frontiera) Siano $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in\mathbb{R}^n$. $\overline{x_0}$ si dice **punto di frontiera** di $A$ se $\overline{x_0}$ è aderente sia ad $A$ che a $\mathbb{r}^n \setminus A$. Cioè

$$\exists\delta > 0\ \text{tale che}\ A \cap B_{\delta}(\overline{x_0})\neq\emptyset\ \text{e}\ (\mathbb{R}^n \setminus A) \cap B_{\delta}(\overline{x_0})\neq\emptyset$$

**Definizione** (frontiera di un insieme) Sia $A\subseteq\mathbb{R}^n$. Chiamiamo **frontiera** di $A$ l'insieme dei punti di frontiera di $A$, denotato con $\partial A$.

1. $\overline{A} = A \cup \partial A = \overset{\circ}{A} \cup \partial A$

**Definizione** (punto di accumulazione) Siano $A\subseteq\mathbb{R}^n$ e $\overline{»}_0\in\mathbb{R}^n$. $\overline{x_0}$ si dice **punto di accumulazione** per $A$ se

$$\forall\delta > 0 (B_{\delta}(\overline{x_0}) \setminus \{\overline{x_0}\}) \cap A \neq \emptyset$$

**Osservazione** Se $\overline{x_0}$ è punto di accumulazione per $A$, allora $\overline{x_0}$ è punto di aderenza di $A$.

**Osservazione** L'insieme dei punti di accumulazione è contenuto in $\overline{A}$ e contiene $\overset{\circ}{A}$.

**Proposizione** Se $A\subseteq\mathbb{R}^n$ è chiuso allora contiene i suoi punti di accumulazione.

**Teorema** (di Bolzano-Weierstrass) Un insieme infinito e limitato in $\mathbb{R}^n$ possiede almeno un punto di accumulazione.

**Definizione** $A\subseteq\mathbb{R}^n$ è limitato se $\exists M > 0\ \text{tale che}\ A \subseteq B_{M}(\overline{0})$.

## Funzioni

**Definizione** (funzione a valori scalari) Sia $A\subseteq\mathbb{R}^n$. $f:A\rightarrow\mathbb{R}$ è un'applicazione che a ogni $\overline{x}\in A$ associa il numero reale $f(\overline{x})\in\mathbb{R}$.

**Definizione** (grafico di $f:A\subseteq\mathbb{R}^n\rightarrow\mathbb{R}$)

$$\mathcal{G}_{f} = \{ (x_1,\dots, x_n, y)\in A\times\mathbb{R}\subseteq\mathbb{R}^{n+1}\ |\ y = f(x_1,\dots, x_n) \}$$

**Definizione** (funzione a valori vettoriali) Sia $A\subseteq\mathbb{R}^n$. $f:A\rightarrow\mathbb{R}^p$ è un'applicazione che a ogni $\overline{x}\in A$ associa il vettore $f(\overline{x}) = (f_1(\overline{x}),\dots, f_p(\overline{x})) \iff f_1,\dots, f_p: A\rightarrow\mathbb{R}$.

**Definizione** (grafico di $f:A\subseteq\mathbb{R}^n\rightarrow\mathbb{R}^p$)

$$\mathcal{G}_{f} = \{ (x_1,\dots, x_n, y_1,\dots, y_p)\in A\times\mathbb{R}^p\subseteq\mathbb{R}^{n+p}\ |\ y_1 = f_1(x_1,\dots, x_n),\dots, y_p = f_p(x_1,\dots, x_n) \}$$

**Definizione** (campo vettoriale) Sia $A\subseteq\mathbb{R}^n$. $f:A\rightarrow\mathbb{R}^n$ è un'applicazione che a ogni $\overline{x}\in A$ associa il vettore $\overline{y} = f(\overline{x}) = (f_1(\overline{x}),\dots, f_n(\overline{x}))$.

## Composizione di funzioni

**Definizione** Siano $f:A\subseteq\mathbb{R}^n\rightarrow\mathbb{R}^p$ e $g:B\subseteq\mathbb{R}^p\rightarrow\mathbb{R}^l$. Sia $f(A)\subseteq B\subseteq\mathbb{R}^p$. Allora

$$g \circ f: A\rightarrow\mathbb{R}^l$$

e partendo da $\overline{x} = (x_1,\dots,x_n)$

$$(g \circ f)(\overline{x}) = g(f(\overline{x})) = g(f_1(\overline{x}),\dots,f_n(\overline{x})) = (g_1(f_1(\overline{x}),\dots,f_p(\overline{x})),\dots,g_l(f_1(\overline{x}),\dots,f_p(\overline{x})))$$

# Limiti

**Definizione** (limite per funzioni a valori scalari) Sia $A\subseteq\mathbb{R}^n$, $f:A\rightarrow\mathbb{R}$, $\overline{x}_0\in\mathbb{R}^n$ punto di accumulazione per $A$, $l\in\mathbb{R}$. Diciamo che

$$\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})} = l$$

se

$$\forall\epsilon > 0\ \exists\delta > 0\ \text{tale che}\ \forall\overline{x}\in A, \overline{x}\neq\overline{x_0}, d(\overline{x},\overline{x_0}) < \delta, |f(\overline{x}) - l| < \epsilon$$

**Definizione** (limite per funzioni a valori vettoriali) Siano $A\subseteq\mathbb{R}^n$, $f=(f_1,...,f_p):A\rightarrow\mathbb{R}^p$, $\overline{x_0}\in\mathbb{R}^n$ punto di accumulazione per $A$, $\overline{l}=(l_1,...,l_p)\in\mathbb{R}^p$. Diciamo che

$$lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=\overline{l} \iff \left\{\begin{matrix}
\lim_{\overline{x}\rightarrow\overline{x_0}}{f_1(\overline{x})}=l_1\\ 
\vdots \\
\lim_{\overline{x}\rightarrow\overline{x_0}}{f_p(\overline{x})}=l_p
\end{matrix}\right.$$

Cioè

$$\forall\epsilon>0\ \ \exists\delta>0\ \ \text{tale che}\ \ d_r(f(\overline{x},\overline{l})<\epsilon\ \ \forall\overline{x}\in A,\ \overline{x}\neq\overline{x_0}\ \ d_n(\overline{x},\overline{x_0})<\delta$$

## Proprietà

1. **Unicità del limite**: il limite, se esiste, è unico

2. **Località del limite**: invece di considerare $f$ in tutto il dominio, posso considerare $f$ definita in $A\cap B_{r}(\overline{x_0})$.

3. **Locale limitatezza di $f$**: se il limite esiste finito e vale $l$, $\exists r>0$ tale che $|f(\overline{x})|\leq M\ \forall\overline{x}\in A\cap B_r(\overline{x_0})$.

**Teorema** (limiti e relazione d'ordine) Siano $A\subseteq\mathbb{R}^n$, $f,g:A\rightarrow\mathbb{R}$, $\overline{x_0}\in\mathbb{R}^n$ punto di accumulazione per $A$, $l,m\in\mathbb{R}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=l$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{g(\overline{x})}=m$. Allora:

1. Se $f(\overline{x}) \leq g(\overline{x})\ \forall\overline{x}\in A$, allora $l<m$.

2. Se $l<m$ allora $\exists r>0\ \text{tale che}\ f(\overline{x})\leq g(\overline{x})\ \forall\overline{x}\in A\cap(B_r(\overline{x})\setminus\{\overline{x_0}\})$.

**Teorema** (proprietà algebriche dei limiti) Siano $A\subseteq\mathbb{R}^n$, $f,g:A\rightarrow\mathbb{R}$, $\overline{x_0}$ punto di accumulazione per $A$. Siano $l,m\in\mathbb{R}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=l$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{g(\overline{x})}=m$. Allora:

1. $\exists\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{[f(\overline{x})\pm g(\overline{x})]}=l\pm m$

2. $\exists\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{[f(\overline{x})\cdot g(\overline{x})]}=l\cdot m$

3. Se $m\neq 0$ allora $\exists\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{\frac{1}{g(\overline{x})}} = \frac{1}{m}$

4. Se $m\neq 0$ allora $\exists\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{\frac{f(\overline{x})}{g(\overline{x})}}=\frac{l}{m}$

**Teorema** (proprietà algebriche degli infinitesimi) Siano $A\subseteq\mathbb{R}^n$, $f,g:A\rightarrow\mathbb{R}$, $\overline{x_0}$ punto di accumulazione per $A$. Siano $l,m\in\mathbb{R}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=0$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{g(\overline{x})}=m$. Allora:

1. Se $\exists M>0$ tale che $|g(\overline{x})|<M\ \forall \overline{x}\in A$, allora $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})g(\overline{x})}=0$

2. Se $f(\overline{x})>0\ \forall\overline{x}\in A$, allora $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{\frac{1}{f(\overline{x})}}=+\infty$

**Definizione** Sia $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in\mathbb{R}^n$ punto di accumulazione per $A$, $f:A\rightarrow\mathbb{R}$. Diciamo che $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=+\infty$ se $\forall\epsilon>0\ \ \exists\delta>0\ \ f(\overline{x})>\epsilon\ \ \forall\overline{x}\in A,\ \ \overline{x}\neq\overline{x_0},\ d(\overline{x},\overline{x_0})<\delta$

**Teorema** (proprietà degli infiniti) Siano $A\subseteq\mathbb{R}^n$, $f,g:A\rightarrow\mathbb{R}$, $\overline{x_0}$ punto di accumulazione per $A$. Siano $l,m\in\mathbb{R}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=+\infty$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{g(\overline{x})}=m$. Allora:

1. Se $\exists M\in\mathbb{R}$ tale che $g(\overline{x})\geq M\ \ \forall\overline{x}\in A$, allora $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})+g(\overline{x})}=+\infty$

2. Se $\exists M>0$ tale che $g(\overline{x})\geq M\ (risp. g(\overline{x})\leq -M)\ \forall\overline{x}\in A$, allora $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})g(\overline{x}}=+\infty (risp. -\infty)$

3. $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{\frac{1}{f(\overline{x})}}=0$

**Teorema** (proprietà algebriche delle funzioni a valori vettoriali) Siano $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in\mathbb{R}^n$ punto di accumulazione per $A$, $f=(f_1,...,f_p),\ g=(g_1,...,g_p):A\rightarrow\mathbb{R}^p$, $\overline{l}=(l_1,...,l_p)\in\mathbb{R}^n$ e $\overline{m}=(m_1,...m_p)\in\mathbb{R}^p$. Se $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=\overline{l}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{g(\overline{x})}=\overline{m}$, allora:

1. $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})+g(\overline{x})}=\overline{l}+\overline{m}$

2. Se $c\in\mathbb{R}$, $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{cf(\overline{x})}=c\overline{l}$

3. $||f(\overline{x})||=\sqrt{f_1^2(\overline{x})+...+f_p^2(\overline{x}}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{||f(\overline{x})||}=||\overline{l}||=\sqrt{l_1^2+...+l_p^2}$

4. $\langle{f(\overline{x}),g(\overline{x})}\rangle=\displaystyle\sum_{i=1}^{p}{f_i(\overline{x})g_i(\overline{x})}\in\mathbb{R}$ e $\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{\langle{f(\overline{x}),g(\overline{x})}\rangle}=\langle{\overline{l},\overline{m}}\rangle$

**Osservazione** L'unicità del limite vale anche per le funzioni a valori vettoriali e deriva dall'unicità dei limiti delle singole componenti.

**Teorema** (unicità del limite per funzioni a valori vettoriali) Siano $A\subseteq\mathbb{R}^n$, $\overline{x}\in\mathbb{R}^n$ punto di accumulazione per $A$, $f:A\rightarrow\mathbb{R}^p$. Allora se $\exists\displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x}}=\overline{l}\in\mathbb{R}^p$, $\overline{l}$ è unico.

## Continuità

**Definizione** (insieme connesso) $A\subseteq\mathbb{R}^n$ è un **insieme connesso** se $!\exists B_1,B_2\subseteq\mathbb{R}^n$ aperti disgiunti tali che $(B_1\cap A)\cup(B_2\cap A)=A$ formino una partizione di $A$.

**Proposizione** $A\subseteq\mathbb{R}$ è connesso $\iff$ $A$ è un intervallo.

**Definizione** (insieme compatto) $A\subseteq\mathbb{R}^n$ è un **insieme compatto** se è limitato e chiuso.

**Osservazione** L'insieme di Cantor $\mathcal{C}$ è compatto. L'insieme $A=\{x=\frac{1}{n}, n\in\mathbb{N}\}$ non è compatto. $A\cup\{0\}$ è compatto.

**Definizione** (funzione continua in un punto) Siano $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in A$, $f:A\rightarrow\mathbb{R}^p$. Diciamo che $f$ è continua in $\overline{x_0}$ se $\forall\epsilon>0\ \ \exists\delta>0$ tale che $d_p(f(\overline{x}),f(\overline{x_0}))<\epsilon\ \ \forall\overline{x}\in A$ tale che $d_n(\overline{x},\overline{x_0})<\delta$.

**Osservazione**

1. Se $\overline{x_0}\in A$ è di accumulazione, allora $f$ è continua in $\overline{x_0} \iff \displaystyle\lim_{\overline{x}\rightarrow\overline{x_0}}{f(\overline{x})}=f(\overline{x_0})$

2. Se $\overline{x_0}\in A$ non è punto di accumulazione per $A$, allora $f$ è banalmente continua in $\overline{x_0}$

**Teorema** (proprietà algebriche delle funzioni continue a valori scalari)
Siano $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in A$, $f,g:A\rightarrow\mathbb{R}$ continue in $\overline{x_0}$. Allora:

1. $f+g$ e $f\cdot g$ sono continue in $\overline{x_0}$

2. Se $g(\overline{x})\neq0$ allora $\frac{1}{g}$ e $\frac{f}{g}$ sono continua in $\overline{x_0}$

**Teorema** (proprietà algebriche delle funzioni continue a valori vettoriali)
Siano $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in A$, $f,g:A\rightarrow\mathbb{R}^p$ funzioni continue in $\overline{x_0}$. Allora:

1. $f+g$ è continua in $\overline{x_0}$

2. $||f(\overline{x})||$ è continua in $\overline{x_0}$

3. $\langle{f(\overline{x}),g(\overline{x})}\rangle$ è continua in $\overline{x_0}$

**Teorema** (continuità della funzione composta) Siano $A\subseteq\mathbb{R}^n$, $B\subseteq\mathbb{R}^p$, $f:A\rightarrow\mathbb{R}^p$ tale che $f(A)\subseteq B$, $\overline{x_0}\in A$, $\overline{y_0}\in\mathbb{R}^p$, $f$ continua in $\overline{x_0}$, $g:B\rightarrow\mathbb{R}^l$ continua in $\overline{y_0}=f(\overline{x_0}$. Allora:

$$g\circ f:A\rightarrow\mathbb{R}^l\ \text{è continua in}\ \overline{x_0}\in A$$

**Teorema** (permanenza del segno) Siano $A\subseteq\mathbb{R}^n$, $\overline{x_0}\in A$, $f:A\rightarrow\mathbb{R}$ continua in $\overline{x_0}$. Se $f(\overline{x_0})>0$ allora $\exists\delta>0$ tale che $\forall\overline{X}\in B_{\delta}(\overline{x_0})\cap A\ \ f(\overline{x})>0$.

**Osservazione** Questo teorema ha senso se $f$ è a valori scalari.

**Teorema** (di Bolzano) Siano $A\subseteq\mathbb{R}^n$ connesso e $f:A\rightarrow\mathbb{R}^p$ continua in $A$. Allora $f(A)\subseteq\mathbb{R}^p$ è connesso.

**Teorema** (di Weierstrass) (1) Siano $A\subseteq\mathbb{R}^n$ compatto e $f:A\rightarrow\mathbb{R}$ continua in $A$. Allora $\exists\displaystyle\min_A{f}$ e $\exists\displaystyle\max_A{f}$.

**Teorema** (di Weierstrass) (2) Siano $A\subseteq\mathbb{R}^n$ compatto e $f:A\rightarrow\mathbb{R}^p$ continua in $A$. Allora $f(A)$ è un insieme compatto in $\mathbb{R}^p$.

# Calcolo differenziale per funzioni a valori scalari

**Definizione** Siano $A\subseteq\mathbb{R}^n$, $f:A\rightarrow\mathbb{R}$ e $\overline{x^0}\in A$. Diciamo che $f$ è derivabile parzialmente rispetto a $x_i$ nel punto $\overline{x^0}$ se esiste finito il limite

$$\lim_{h\rightarrow 0}{\frac{f(x_1^0,...,x_i^0+h,...,x_n^0)-f(x_1^0,...,x_n^0)}{h}}=\frac{\partial f}{\partial x_i}$$

**Definizione** (gradiente) Siano $A\subseteq\mathbb{R}^n$ aperto, $f:A\rightarrow\mathbb{R}$ derivabile in $\overline{x_0}\in A$. Si dice **gradiente** di $f$ in $\overline{x_0}$ il vettore

$$Df(\overline{x_0})=\nabla{f(\overline{x})}=(\frac{\partial f}{\partial x_1}(\overline{x_0}),...,\frac{\partial f}{\partial x_n}(\overline{x_0}))$$

**Osservazione** Siano $A\subseteq\mathbb{R}^n$ aperto, $f:A\rightarrow\mathbb{R}$ derivabile in $\overline{x_0}\in A$. Se esiste il gradiente$\nabla{f(\overline{x_0})}$, allora non è detto che $f$ sia continua in $\overline{x_0}$.

**Teorema** (condizioni sufficienti per l'esistenza del piano tangente al grafico di una funzione)
Siano $A\subseteq\mathbb{R}^n$ aperto, $f:A\rightarrow\mathbb{R}$ e $\overline{x^0}\in A$ e supponiamo esistano continue le derivate parziali di $f$ in $\overline{x^0}$. Allora

$$y=f(\overline{x^0})+\langle{\nabla{f(\overline{x^0})}, \overline{x}-\overline{x^0}}\rangle=f(\overline{x^0})+\sum_{i=1}^{n}{\frac{\partial f}{\partial x_i}(\overline{x_i}-\overline{x^0_i})}$$

**Teorema** (formula di Taylor al primo ordine) Siano $A\subseteq\mathbb{R}^n$ aperto, $f\in\mathcal{C}^{(1)}(A,\mathbb{R})$, $\overline{x_0}\in A$, $\overline{h}=(h_1,...,h_n)\in\mathbb{R}^n$. Allora

$$f(\overline{x_0}+\overline{h}) = f(\overline{x_0}) + \langle{\nabla{f(\overline{x_0})}, \overline{h}}\rangle + o(||\overline{h}||)\ \ \ ,\ \ \ ||\overline{h}||\rightarrow{0}$$

**Definizione** (derivata direzionale) Siano $A\subseteq\mathbb{R}^n$ aperto, $f:A\rightarrow\mathbb{R}$, $\overline\lambda=(\lambda_1,...,\lambda_n)\in\mathbb{R}^n$ con $||\overline\lambda||=1$. Chiameremo **derivata direzionale** di $f$ nella direzione di $\overline\lambda$ il limite, se esiste finito, seguente

$$\lim_{h\rightarrow{0}}{\frac{f(\overline{x}+h\overline{\lambda}) - f(\overline{x})}{h}}$$

Essa è indicata con i simboli $\frac{\partial f}{\partial\lambda}(\overline{x})$ o $D_{\lambda}f(\overline{x})$.

**Teorema** Siano $A\subseteq\mathbb{R}^n$ aperto, $f\in\mathcal{C}^{(1)}(A,\mathbb{R}^n)$, $\overline{x}\in A$. Allora $\forall\overline\lambda\in\mathbb{R}^n$ con $||\overline\lambda||=1$ si ha:

1. $\frac{\partial f}{\partial\overline\lambda}=\langle{\nabla{f(\overline{x})},\overline\lambda}\rangle$

2. Preso $\tilde\lambda=\frac{\nabla{f(\overline{x})}}{||\nabla{f(\overline{x})}||}$ si ha $\frac{\partial f}{\partial\overline\lambda}=\langle{\nabla{f(\overline{x})},\tilde\lambda}\rangle=\frac{||\nabla{f(\overline{x})}||^2}{||\nabla{f(\overline{x})}||}=\nabla{f(\overline{x})}$

**Definizione** (derivate parziali di ordine superiore) Sia $A\subseteq\mathbb{R}^n$ aperto, $f:A\rightarrow\mathbb{R}$. Se $\frac{\partial f}{\partial x_i}$ è derivabile in $A$, chiamiamo derivata parziale seconda di $f$ rispetto a $x_ix_j$, se esiste finito, il  nel punto $\overline{x_0}\in A$

$$\frac{\partial^2 f}{\partial x_ix_j}=\lim_{h\rightarrow{0}}{\frac{\frac{\partial f}{\partial x_i}(\overline{x_0}+h\hat{e}_j) - \frac{\partial f}{\partial x_i}(\overline{x_0})}{h}}$$

**Definizione** (matrice hessiana) Supponiamo che $f:A\rightarrow\mathbb{R}$, $A\subseteq\mathbb{R}^n$ aperto, possegga tutte le derivate parziali al secondo ordine incluso in $\overline{x_0}\in A$. Allora chiamiamo **matrice hessiana** di $f$ in $\overline{x_0}$ la matrice

$$H_f(\overline{x_0})=
\begin{bmatrix}
\nabla{\frac{\partial f}{\partial x_1}(\overline{x_0})} & \vdots & \nabla{\frac{\partial f}{\partial x_n}(\overline{x_0})}
\end{bmatrix}=
\begin{bmatrix}
\frac{\partial^2 f}{\partial x_1^2}(\overline{x_0}) & \dots & \frac{\partial^2 f}{\partial x_1x_n}(\overline{x_0})\\ 
\vdots & \ddots & \vdots\\ 
\frac{\partial^2 f}{\partial x_nx_1}(\overline{x_0}) & \dots & \frac{\partial^2 f}{\partial x_n^2}(\overline{x_0})
\end{bmatrix}$$

**Teorema** (lemma di Schwarz) Siano $A\subseteq\mathbb{R}^n$ aperto e $f\in\mathcal{C}^{(2)}(A,\mathbb{R})$ tale che $\frac{\partial^2 f}{\partial x_ix_j}$ e $\frac{\partial^2 f}{\partial x_jx_i}$ siano continue in $\overline{x_0}\in A$. Allora

$$\frac{\partial^2 f}{\partial x_ix_j}(\overline{x_0})=\frac{\partial^2 f}{\partial x_jx_i}(\overline{x_0})$$

cioè $H_f(\overline{x_0})$ è simmetrica.

**Teorema** (formula di Taylor al secondo ordine) Siano $A\subseteq\mathbb{R}^n$ aperto, $f:A\rightarrow\mathbb{R}$ con derivate parziali continue fino al secondo ordine incluso in $\overline{x_0}\in A$. Allora:

$$f(\overline{x_0}+\overline{h})=f(\overline{x_0})+\langle{\nabla{f(\overline{x_0})},\overline{h}}\rangle+\frac{1}{2}\langle{H_f(\overline{x_0})\overline{h},\overline{h}}\rangle+o(||\overline{h}||^2)\ \ \ ,\ \ \ ||h||\rightarrow{0}$$

**Definizione** (massimo e minimo locale) Siano $A\subseteq\mathbb{R}^n$, $f:A\rightarrow\mathbb{R}$. Diciamo che $\overline{x_0}\in A$ è un punto di **massimo locale** (risp. di **minimo locale**) se $\exists\delta>0$ tale che

$$f(\overline{x_0})\geq f(\overline{x})\ \forall\overline{x}\in A\cap B_{\delta}(\overline{x_0})\ \ \text{(risp.}\ f(\overline{x_0})\leq f(\overline{x})\text{)}$$

**Teorema** (di Fermat) Siano $f:A\rightarrow\mathbb{R}$, $A\subseteq\mathbb{R}^n$ aperto, $\overline{x_0}\in A$ punto di massimo o minimo locale. Se $\exists\nabla{f(\overline{x_0})}$ allora $\nabla{f(\overline{x_0})}=0$.

**Dimostrazione** Verifichiamo $\frac{\partial f}{\partial x_1}(\overline{x_0})=0$ con $\overline{x_0}$ punto di massimo.

$$\frac{\partial f}{\partial x_1}(\overline{x_0})=\lim_{h\rightarrow{0}}{\frac{f(\overline{x_0}+h\hat{e_1})-f(\overline{x_0}}{h}}$$

Chiamiamo $F(h)=f(\overline{x_0}+h\hat{e_1})$, $F(0)=F(\overline{x_0})$. Siccome $\overline{x_0}$ è un punto di massimo locale, $\exists\delta>0$ tale che

$$F(h)\leq F(0)\ \ \forall |h|<\delta$$

$F:]-\delta,\delta[\rightarrow\mathbb{R}$ ha quindi massimo locale in $h=0$. Inoltre $F(h)$ è derivabile in $0$:

$$F'(0)=\lim_{h\rightarrow 0}{\frac{F(h)-F(0)}{h}}=\frac{\partial f}{\partial x_1}(\overline{x_0})$$

e $\frac{\partial f}{\partial x_1}(\overline{x_0})=0$ per il teorema di Fermat in $n=1$. In maniera analoga si verificano le altre componenti.

**Osservazione** E' una condizione necessaria ma non sufficiente.

**Definizione** (forme quadratiche) Sia $A$ una matrice $nxn$. Una **forma quadratica** è un'applicazione $q_A:\mathbb{R}^n\rightarrow\mathbb{R}$

$$q_A(\overline{h})=\sum_{i,j=1}^{n}{a_{ij}h_1h_j}$$

$q_A$ è:

- definita positiva se $q_A(\overline{h})>0\ \ \forall\overline{h}\neq{0}$

- definita negativa se $q_A(\overline{h})<0\ \ \forall\overline{h}\neq{0}$

- semidefinita positiva se $q_A(\overline{h})\geq{0}\ \ \forall\overline{h}\in\mathbb{R}^n$

- semidefinita negativa se $q_A(\overline{h})\leq{0}\ \ \forall\overline{h}\in\mathbb{R}^n$

- indefinita se $\exists \overline{h},\overline{k}\in\mathbb{R}^n$ tali che $q_A(\overline{h})<0<q_A(\overline{k})$

**Osservazione** $\forall A\in M_{nxn}\ \ !\exists q_A$

**Osservazione** Presa una forma quadratica $q_A$ possiamo associarle infinite matrici, ma una sola è simmetrica.

**Teorema** (forme quadratiche e autovalori) Sia $A\in M_{nxn}$ simmetrica, $\lambda_1,...,lambda_n\in\mathbb{R}$ i suoi autovalori e $q_A$ la forma quadratica associata ad $A$. Allora $q_A$ è:

- definita positiva $\iff\ \lambda_1,...,\lambda_n>0$

- definita negativa $\iff\ \lambda_1,...,\lambda_n<0$

- semidefinita positiva in senso proprio $\iff\ \lambda_1,...,\lambda_n\geq{0}$ e $\exists\lambda_i,\lambda_j$ tali che $\lambda_i=0$ e $\lambda_j>0$

- semidefinita negativa in senso proprio $\iff\ \lambda_1,...,\lambda_n\leq{0}$ e $\exists\lambda_i,\lambda_j$ tali che $\lambda_i=0$ e $\lambda_j<0$

- indefinita $\iff\ \exists\lambda,\mu$ autovalori tali che $\lambda<0<\mu$

**Teorema** (criterio di Sylvester per la matrici $2x2$) Sia $A\in M_{2x2}$ simmetrica. Allora:

- $A$ è definita $\iff\ \det(A)>0$

- $A$ è semidefinita $\iff\ \det(A)=0$

- $A$ è indefinita $\iff\ \det(A)<0$

- $A$ è definita positiva $\iff\ \det(A)>0$ e $a_{11}>0$

- $A$ è definita negativa $\iff\ \det(A)<0$ e $a_{11}<0$

- $A$ è semidefinita positiva $\iff\ \det(A)=0$ e $a_{11},a_{22}>0$

- $A$ è semidefinita negativa $\iff\ \det(A)=0$ e $a_{11},a_{22}<0$

**Teorema** (criterio di Sylvester per le matrici $nxn$) Sia $A\in M_{nxn}$ simmetrica. Allora:

- $A$ è definita positiva $\iff\ \det(A^{(k)})>0\ \forall k=1,...,n$

- $A$ è definita negativa $\iff\ (-1)^k\dot\det(A^{(k)})>0\ \forall k=1,...,n$

- Se $\det(A)^{(k)}\neq{0}$ ma non rispetta le regole di segno precedenti, allora $A$ è indefinita.

## Punti critici

**Teorema** Sia $A\subseteq\mathbb{R}^n$ aperto, $f\in\mathcal{C}^{(2)}(A,\mathbb{R})$ e $\overline{x_0}\in A$. Allora:

1. se $\overline{x_0}$ è un punto di minimo locale, allora $\nabla{f(\overline{x_0}}=\overline{0}$ e $H_f(\overline{x_0})$ è semidefinita positiva

2. se $\overline{x_0}$ è un punto di massimo locale, allora $\nabla{f(\overline{x_0}}=\overline{0}$ e $H_f(\overline{x_0})$ è semidefinita negativa

3. se $\nabla{f(\overline{x_0}}=\overline{0}$ e $H_f(\overline{x_0})$ è definita positiva, allora $\overline{x_0}$ è un punto di minimo locale

4. se $\nabla{f(\overline{x_0}}=\overline{0}$ e $H_f(\overline{x_0})$ è definita negativa, allora $\overline{x_0}$ è un punto di massimo locale

5. se $\nabla{f(\overline{x_0}}=\overline{0}$ e $H_f(\overline{x_0})$ è indefinita, allora $\overline{x_0}$ è un punto di sella

**Dimostrazione** (4) Usiamo la formula di Taylor al secondo ordine

$$f(\overline{x_0}+\overline{h})=f(\overline{x_0})+\langle{\nabla{f(\overline{x_0})},\overline{h}}\rangle+\frac{1}{2}\langle{H_f(\overline{x_0})\overline{h},\overline{h}}\rangle+o(||\overline{h}||^2)\ \ \ ,\ \ \ ||h||\rightarrow{0}$$

$\langle{\nabla{f(\overline{x_0})},\overline{h}}\rangle=0$ per ipotesi e $\langle{H_f(\overline{x_0})\overline{h},\overline{h}}\rangle=q_A(\overline{h})$. Sia quindi $M<0$ l'autovalore massimo di $H_f(\overline{x_0})$

$$f(\overline{x_0}+\overline{h})-f(\overline{x_0})=\frac{1}{2}\langle{H_f(\overline{x_0})\overline{h},\overline{h}}\rangle+o(||\overline{h}||^2)\leq M||h||^2+o(||\overline{h}||^2)\ \ \ ,\ \ \ ||h||\rightarrow{0}$$

Quindi $f(\overline{x_0}+\overline{h})-f(\overline{x_0})<0\ \implies\ \overline{x_0}$ è un punto di massimo.

**Osservazione** Se $\nabla{f(\overline{x_0})}=0$ e $H_f(\overline{x_0})$ è semidefinita positiva in senso proprio sicuramente $\overline{x_0}$ non potrà essere di massimo locale. Sarà quindi un punto di minimo locale o sella.

# Calcolo differenziale per funzioni a valori vettoriali

**Definizione** Sia $A\subseteq\mathbb{R}^n$, $f:A\rightarrow\mathbb{R}^p$. $f$ si dice derivabile parzialmente in $\overline{x_0}\in A$ se ogni componente $f_i$ è derivabile parzialmente in $\overline{x_0}$. In tal caso si definisce **matrice jacobiana** di $f$ in $\overline{x_0}$ la matrice $pxn$ i cui elementi della riga $r^{ma}$ sono le componenti del gradiente di $f_r$.

$$J_f(\overline{x_0})=
\begin{bmatrix}
\frac{\partial f_1}{\partial x_1} & \dots & \frac{\partial f_1}{\partial x_n}\\ 
\vdots & \ddots & \vdots\\ 
\frac{\partial f_p}{\partial x_1} & \dots & \frac{\partial f_n}{\partial x_n}
\end{bmatrix}=
\begin{bmatrix}
\nabla{f_1(\overline{x_0})}\\
\vdots\\
\nabla{f_p(\overline{x_0})}
\end{bmatrix}$$

**Osservazione** Se $p=1$, $J_f(\overline{x_0})=\nabla{f(\overline{x_0})}$.

**Osservazione** Se $f\in\mathcal{C}^{(1)}(A,\mathbb{R}^p)$, usando la formula di Taylor per le componenti $f_i,\ i=1,...,p$, abbiamo per ogni $\overline{x_0}\in A$ fissato

$$\begin{bmatrix}
f_1(\overline{x_0}+\overline{h})\\
\vdots\\
f_p(\overline{x_0}+\overline{h})
\end{bmatrix}=
\begin{bmatrix}
f_1(\overline{x_0})+\langle{\nabla{f_1(\overline{x_0})},\overline{h}}\rangle+o(||h||)\\
\vdots\\
f_p(\overline{x_0})+\langle{\nabla{f_p(\overline{x_0})},\overline{h}}\rangle+o(||h||)
\end{bmatrix}$$

cioè

$$f(\overline{x_0}+\overline{h})=f(\overline{x_0})+\langle{J_f(\overline{x_0}),\overline{h}}\rangle+o(||h||)\ \ \ ,\ \ \ ||\overline{h}||\rightarrow{0}$$

## Composizione di funzioni

**Teorema** (derivazione di funzione composta) Siano $A\subseteq\mathbb{R}^n$, $B\subseteq\mathbb{R}^p$, $f\in\mathcal{C}^{(1)}(A,B)$, $g\in\mathcal{C}^{(1)}(B,\mathbb{R}^k)$. Allora $g\circ f\in\mathcal{C}^{(1)}(A,\mathbb{R}^k)$ e

$$J_{g\circ f}(\overline{x})=J_g(f(\overline{x}))\cdot J_f(\overline{x})\ \ \ ,\ \ \ \forall\overline{x}\in A$$

ovvero

$$J_{g\circ f}(\overline{x})=\big[\langle{\nabla{g_i(f(\overline{x}))}, \frac{\partial f}{\partial x_j}(\overline{x})}\rangle\big]_{j=1,...,n}^{i=1,...,k}$$

# Successioni numeriche

**Definizione** È detta **successione** una applicazione $f:\mathbb{N}\rightarrow\mathbb{R}\ (o\ \mathbb{C})$. I valori di questa successione $x_n = f(n)$ sono detti elementi della successione e scriveremo $\{x_n\}_{n\in\mathbb{N}}$ sia i valori della successione che la successione stessa.

**Definizione** (limiti di successioni reali) Sian $\{x_n\}_{n\in\mathbb{N}}$ una successione reali, $l\in\mathbb{R}$. Diremo che $\{x_n\}_{n\in\mathbb{N}}$ **converge** a $l$ e scriveremo $x_n\underset{n\rightarrow +\infty}{\rightarrow}l$ o $\displaystyle\lim_{n\rightarrow+\infty}{x_n}=l$ se

$$\forall\epsilon > 0\ \ \exists\nu(\epsilon)\in\mathbb{N}\ \text{tale che}\ |x_n-l| \leq \epsilon\ \ \forall n\geq\nu(\epsilon)$$

In questo caso diremo che la successione è regolare o convergente

**Osservazione** La definizione precedente vale anche nel caso in cui $\{z\}_{n\in\mathbb{N}}$ sia una successione a valori complessi.

**Definizione** Diremo che una successione reale $\{x_n\}_{n\in\mathbb{N}}$ **diverge** positivamente (risp. negativamente) se

$$\forall k > 0\ \ \exists\nu(k)\in\mathbb{N}\ \text{tale che}\ x_n \geq k\ (\text{risp.}\ x_n \leq -k)\ \ \forall n \geq \nu(k)$$

**Definizione** Una successione reale che non sia né convergente né divergente la diremo **oscillante**.

**Definizione** Una successione $\{x_n\}_{n\in\mathbb{N}}$ è detta **limitata** se $\exists M > 0$ tale che $|x_n| < M\ \ \forall n\in\mathbb{N}$.

**Osservazione** Questa definizione ha senso anche per successioni in $\mathbb{C}$.

**Teorema** Siano $\{x_n\}_{n\in\mathbb{N}}$ e $\{y_n\}_{n\in\mathbb{N}}$ successioni regolari con $x_n\rightarrow l$ e $y_n\rightarrow m$, $l, m\in\mathbb{R}$. Allora:

1. $x_n + y_n \rightarrow l + m$

2. $x_n \cdot y_n \rightarrow l \cdot m$

3. se $m \neq 0$ e $y_n \neq 0\ \ \forall n\in\mathbb{N}$, allora $\frac{x_n}{y_n} = \frac{l}{m}$

4. se $x_n \leq y_n\ \ \forall n\in\mathbb{N}$ (è sufficiente definitivamente, ossia da un certo indice in poi), allora $l \leq m$

5. se $l < m$, allora $x_n < y_m$ definitivamente

**Teorema** (del confronto) Siano $\{x_n\}_{n\in\mathbb{N}}$, $\{y_n\}_{n\in\mathbb{N}}$, $\{z_n\}_{n\in\mathbb{N}}$ successioni reali e supponiamo che $x_n, y_n \rightarrow l\in\mathbb{R}\cup\{+\infty\}$. Allora se $x_n \leq z_n \leq y_n\ \ \forall n\in\mathbb{N}$, si ha $z_n \rightarrow l$.

**Teorema** Siano $\{x_n\}_{n\in\mathbb{N}}$, $\{y_n\}_{n\in\mathbb{N}}$ successioni reali e supponiamo $\{x_n\}_{n\in\mathbb{N}}$ infinitesima, cioè $x_n \rightarrow  0$. Allora:

1. se $\{y_n\}_{n\in\mathbb{N}}$ è limitata, $x_n \cdot y_n \rightarrow 0$

2. se $\{x_n\}_{n\in\mathbb{N}} > 0$ (risp. $\{x_n\}_{n\in\mathbb{N}} < 0$) $\forall n\in\mathbb{N}$, allora $\frac{1}{x_n} \rightarrow +\infty$ (risp. $x_n \rightarrow -\infty$)

**Teorema** (successioni divergenti) Siano $\{x_n\}_{n\in\mathbb{N}}$, $\{y_n\}_{n\in\mathbb{N}}$ successioni reali. Allora se $x_n \rightarrow +\infty$

1. se $\exists M \in \mathbb{R}$ tale che $y_n \geq M\ \ \forall n\in\mathbb{N}$, $x_n + y_n \rightarrow +\infty$

2. se $\exists M > 0$ tale che $y_n > M\ \ \forall n\in\mathbb{N}$, $x_n \cdot y_n \rightarrow +\infty$

3. se $\exists M < 0$ tale che $y_n < M\ \ \forall n\in\mathbb{N}$, $x_n \cdot y_n \rightarrow -\infty$

4. $\frac{1}{x_n} \rightarrow 0$

Analogamente per $x_n \rightarrow -\infty$

1. se $\exists M \in \mathbb{R}$ tale che $y_n \leq M\ \ \forall n\in\mathbb{N}$, $x_n + y_n \rightarrow -\infty$

2. se $\exists M > 0$ tale che $y_n > M\ \ \forall n\in\mathbb{N}$, $x_n \cdot y_n \rightarrow -\infty$

3. se $\exists M < 0$ tale che $y_n < M\ \ \forall n\in\mathbb{N}$, $x_n \cdot y_n \rightarrow +\infty$

4. $\frac{1}{x_n} \rightarrow 0$

**Teorema** (criterio di Cauchy) Sia $\{x_n\}_{n\in\mathbb{N}}$ una successione reale (o a valori complessi). Allora essa è regolare se e solo se

$$\forall\epsilon > 0\ \ \exists\nu(\epsilon)\in\mathbb{N}\ \text{tale che}\ |x_n - x_m|\leq\epsilon\ \ \forall n,m \geq \nu(\epsilon)$$

**Osservazione** È una condizione necessaria e sufficiente per la convergenza.

**Definizione** (successioni monotone) La successione numerica $\{x_n\}_{n\in\mathbb{N}}$ è **crescente** (risp. decrescente) se

$$n < m \implies x_n \leq x_m\ (\text{risp.}\ x_n \geq x_m)$$

La successione $\{x_n\}_{n\in\mathbb{N}}$ è strettamente crescente (risp. strettamente decrescente) se

$$n < m \implies x_n < x_m (\text{risp.}\ x_n > x_m)$$

**Definizione** (successione estratta) Sia $\{x_n\}_{n\in\mathbb{N}}$ una successione. Se $h \rightarrow \sigma(h)$ è una successione strettamente crescente da $\mathbb{N}*$ in $\mathbb{N}*$, la successione $\{x_{\sigma(h)}\}_{n\in\mathbb{N}}$ si dice **successione estratta** da $\{x_n\}_{n\in\mathbb{N}}$.

**Proposizione** Ogni successione $\{x_n\}_{n\in\mathbb{N}}$ monotona ha limite finito o infinito.

1. Nel caso di successioni crescenti si ha $\displaystyle\lim_{n \rightarrow +\infty}{x_n} = \sup_n{x_n}$. In particolare tale limite è finito se la successione è limitata, altrimenti la successione diverge positivamente.

2. Nel caso di successioni decrescenti si ha $\displaystyle\lim_{n \rightarrow +\infty}{x_n} = \inf_n{x_n}$. In particolare tale limite è finito se la successione è limitata, altrimenti la successione diverge negativamente.

**Definizione** (numero di Eulero) Il numero di Eulero è definito come 

$$lim_{n \rightarrow +\infty}{(1 + \frac{1}{n})^n} = e$$

**Osservazione** Si può dimostrare che:

1. $x_n = (1 + \frac{1}{n})^n$ è crescente

2. $x_n$ è limitata superiormente (in particolare $2 < x_n < 3\ \ \forall n\in\mathbb{N}$)

**Dimostrazione**

$$(1 + \frac{1}{n})^n = \sum_{h = 0}^{n}{\binom{n}{h}(\frac{1}{n})^h \cdot 1^{n - h}} =$$

$$= \sum_{h = 0}^{n}{\frac{n!}{h!(n - h)!} \cdot \frac{1}{n^h}} =$$

$$= \sum_{h = 0}^{n}{\frac{1}{h!} \cdot (1 - \frac{1}{n}) \cdot (1 - \frac{2}{n})} \cdot ... \cdot (1 - \frac{h -1}{n}) <$$

$$< \sum_{h = 0}^{n}{\frac{1}{h!}} < 1 + \sum_{h = 1}^{n}{\frac{1}{2^{h - 1}}} = \sum_{h = 0}^{n - 1}{\frac{1}{2^h}} = 1 + \frac{2^n - 1}{2^{n - 1}}$$

$$= 1 + 2 - \frac{1}{2^{n-1}} < 3$$

## Successioni di funzioni reali

**Definizione** (limite puntuale) Sia $I \subseteq\mathbb{R}$ intervallo, $g_n, g: I \rightarrow \mathbb{R}$, $n\in\mathbb{N}$. Diremo che $\{g_n\}_{n\in\mathbb{N}}$ converge puntualmente a $g$ se $\forall x \in I\ \ g_n(x)$ converge a $g(x)$. La funzione $g$ è detta limite puntuale della successione $\{g_n\}_{n\in\mathbb{N}}$. In modo equivalente, $g$ è **limite puntuale** se fissato $x$, $\forall\epsilon > 0\ \ \forall x \in I\ \ \exists \nu(\epsilon, x)\in\mathbb{N}$ tale che $|g_n(x) - g(x)| < \epsilon\ \ \forall n \geq \nu(\epsilon, x)$ .

**Definizione** (limite uniforme) Diremo che $\{g_n\}_{n\in\mathbb{N}}$ converge uniformemente a $g$ se detta $\alpha_n = \sup{|g_n(x) - g(x)|}$, si ha che $\alpha \rightarrow 0$ per $n \rightarrow +\infty$. Si scrive

$$lim_{n \rightarrow +\infty}{\sup_I{|g_n(x) - g(x)|}} = 0$$

Equivalentemente:

$$\forall \epsilon > 0\ \ \forall \nu(\epsilon) \in \mathbb{N}\ \text{tale che}\ |g_n(x) - g(x)| \leq \epsilon\ \ \forall n \geq \nu(\epsilon)\ \text{e}\ \forall x \in I$$

La funzione $g$ è detta **limite uniforme**.

**Osservazione** Ovviamente se $g_n \overset{\text{unif.}}{\rightarrow} g$ allora $g_n \overset{\text{punt.}}{\rightarrow} g$. Il contrario è falso.

**Teorema** Sia $\{f_n\}_{n\in\mathbb{N}}$ una successione di funzione $I \rightarrow \mathbb{R}$, $I$ intervallo. Valgono questi fatti:

1. sia $f:I\rightarrow\mathbb{R}$ il limite uniforme di $\{f_n\}_{n\in\mathbb{N}}$. Allora se $f_n$ sono continue in $x_0 \in I$ anche $f$ è continua in $x_0$. Equivalentemente
    
    $$\lim_{x \rightarrow x_0}{\lim_{n \rightarrow +\infty}{f_n(x)}} = \lim_{n \rightarrow +\infty}{\lim_{x \rightarrow x_0}{f_n(x)}}$$
    
2. se $\{f_n\}_{n\in\mathbb{N}}$ sono derivabili in $I$, $\{f_n{x_0}\}_{n\in\mathbb{N}}$ converge per qualche $x_0 \in I$ e $\{f'_n\}_{n\in\mathbb{N}}$ converge uniformemente in $I$, allora:

    I. $\{f_n\}_{n\in\mathbb{N}}$ converge puntualmente in $I$ (e in ogni $[a, b] \subseteq I$)

    II. detta $f$ il limite puntuale di $\{f_n\}_{n\in\mathbb{N}}$, essa risulta derivabile e $f'$ è il limite uniforme di $\{f'_n\}_{n\in\mathbb{N}}$ e si scrive
    
    $$D(\lim_{n \rightarrow +\infty}{f_n(x)}) = \lim_{n \rightarrow +\infty}{[D(f_n(x))]}\ \ \forall x \in I$$
    
3. se $I = [a, b]$, $f_n$ è integrabile secondo Riemann e $f:I \rightarrow \mathbb{R}$ è il limite uniforme di $\{f_n\}_{n\in\mathbb{N}}$, allora

    I. $f$ è integrabile secondo Riemann

    II. $\displaystyle\int_{a}^{b}{f(x)dx} = \int_{a}^{b}{\lim_{n \rightarrow +\infty}{f_n(x)dx}} = \lim_{n \rightarrow +\infty}{\int_{a}^{b}{f_n(x)dx}}$
    
**Dimostrazione** (2.II)

$$f_n(x) - f_m(x) = F_n(x_0) - f_m(x_0) + \int_{x_0}^{x}{|f'_n(t) - f'_m(t)|dt}$$

quindi

$$|f_n(x) - f_m(x)| \leq |f_n(x_0) - f_m(x_0)| + \int_{x_0}^{x}{|f'_n(t) - f'_m(t)|dt} \leq \epsilon + \sup_{[x_0, x]}{|f'_n - f'_m|} \cdot |x_0 - x|$$

se $n, m >> 0$.

# Serie numeriche

**Definizione** Sia $\{x_n\}_{n\in\mathbb{N}}$ una successione in $\mathbb{R}$, poniamo

$$S_n = \sum_{k = 0}^{n}{x_k}$$

$\{S_n\}_{n\in\mathbb{N}}$ è detta **successione delle somme parziali** e anche serie di termine generale di $x_n$ e si indica con

$$\sum_{k = 0}^{+\infty}{x_k}$$

**Definizione** Se esiste finito il $\displaystyle\lim_{n\ rightarrow +\infty}{S_n}$ diremo che la serie $\displaystyle\sum_{k = 0}^{+\infty}{x_k}$ è convergente e scriviamo

$$\sum_{k = 0}^{+\infty}{\lim_{n \rightarrow +\infty}{S_n}} = S$$

**Definizione** Se $\displaystyle\lim_{n \rightarrow +\infty}{S_n} = \pm\infty$ diremo che la serie **diverge** positivamente (o negativamente).

**Definizione** Se $S_n$ non ha limite diremo che la serie è **oscillante**.

**Osservazione** Data una successione $\{x_n\}_{n\in\mathbb{N}}$ in $\mathbb{R}$, se in essa si alterano un numero finito di termini il limite non cambia. Ciò non è vero per la serie associata.

**Definizione** È detta **telescopica** una serie $\displaystyle\sum_{k = 0}^{+\infty}{x_k}$ con $x_k = a_{k+1} - a_k$ per una certa successione $\{a_k\}_{k\in\mathbb{N}}$. In questo caso

$$S_n = x_0 + ... + x_n = (a_1 - a_0) + ... + (a_n - a_{n-1}) = a_{n+1} - a_n$$

$$S = \lim_{n \rightarrow +\infty}{S_n} = \lim_{n \rightarrow +\infty}{(a_{n+1} - a_n)}$$

**Teorema** (criterio di Cauchy) $\displaystyle\sum_{k = 0}^{+\infty}{x_k}$ converge per definizione se converge la successione $\{S_n\}_{n\in\mathbb{N}}$, quindi

$$\displaystyle\sum_{k = 0}^{+\infty}{x_k}\ \text{converge} \iff \forall\epsilon > 0\ \ \exists\nu(\epsilon)\in\mathbb{N}\ \text{tale che}\ |S_n - S_m| \leq \epsilon\ \ \forall n, m \geq \nu(\epsilon)$$

**Teorema** (condizione necessaria di convergenza) Se la serie $\displaystyle\sum_{k = 0}^{+\infty}{x_k}$ è convergente, allora $x_k \rightarrow 0$.

**Definizione** (serie geometrica) Si dice **serie geometrica** di ragione $x$ la serie

$$\sum_{k = 0}^{+\infty}{x^k},\ \ x\in\mathbb{R}$$

- $\{x^n\}_{n\in\mathbb{N}}$ è infinitesima $\iff$ $|x| < 1$

- se $x \geq 1$ la serie diverge positivamente

- se $x \leq -1$ la serie è oscillante

- se $|x| < 1$ allora

    $$S_n = 1 + x + ... + x^n = \frac{1 - x^{n + 1}}{1 - x}$$

    e quindi
    
    $$S = \lim_{n \rightarrow +\infty}{S_n} = \sum_{n = 0}^{+\infty}{\frac{1}{1 - x}}$$
    
**Definizione** (serie armonica) $\displaystyle\sum_{n = 1}^{+\infty}{\frac{1}{n}}$ è detta **serie armonica** e diverge positivamente.

**Dimostrazione** Applichiamo il criterio di Cauchy

$$S_{2n} - S_n = \frac{1}{n + 1} + \frac{1}{n + 2} + ... + \frac{1}{2n} \geq \frac{1}{2n} \cdot n = \frac{1}{2}$$

Quindi la serie non converge.

**Definizione** (convergenza assoluta) Sia $\displaystyle\sum_{k = 0}^{+\infty}{x_k}$ una serie numerica. Diremo che essa è assolutamente convergente se è convergente la serie dei suoi valori assoluti.

**Teorema** Se $\displaystyle\sum_{k = 0}^{+\infty}{x_k}$ converge assolutamente, allora converge semplicemente.

**Dimostrazione** Basta osservare che

$$|S_n - S_m| = |x_{n+1} + ... + x_{n}| \leq |x_{n+1}| + ... + |x_{n}| = |T_n - T_m|$$

dove $\{T_n\}_{n\in\mathbb{N}}$ è la successione delle somme parziali di $\displaystyle\sum_{n = 0}^{+\infty}{|x_n|}$. Se $\displaystyle\sum_{n = 0}^{+\infty}{x_n}$ converge, allora $\{T_n\}_{n\in\mathbb{N}}$ soddisfa Cauchy, quindi $|T_n - T-m| \leq \epsilon$ e quindi $|S_n - S_m| \leq \epsilon$.

## Criteri di convergenza per serie a termini positivi

**Teorema** (criterio del confronto) Siano $\{a_n\}_{n\in\mathbb{N}}$ e $\{b_n\}_{n\in\mathbb{N}}$ due successioni a termini non negativi e supponiamo che $a_n \leq b_n$ definitivamente. Allora:

1. se $\displaystyle\sum_{n = 0}^{+\infty}{b_n}$ converge, allora $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ converge

2. se $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ diverge, allora $\displaystyle\sum_{n = 0}^{+\infty}{b_n}$ diverge

**Teorema** (criterio del rapporto) Siano $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ una serie a termini non negativi (a_n \neq 0 definitivamente) e $\displaystyle l = \lim_{n \rightarrow +\infty}{\frac{a_{n+1}}{a_n}} \in\mathbb{R}\cup\{+\infty\}$. Allora:

1. se $l < 1$ allora $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ converge

2. se $l > 1$ allora $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ diverge

3. se $l = 1$ allora non si può concludere nulla

**Teorema** (criterio della radice) Siano $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ una serie a termini non negativi e $\displaystyle l = \lim_{n \rightarrow +\infty}{\sqrt[n]{a_n}}$. Allora

1. se $l < 1$ allora $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ converge

2. se $l > 1$ allora $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ diverge

3. se $l = 1$ allora non si può concludere nulla

**Teorema** (criterio asintotico) Siano $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ e $\displaystyle\sum_{n = 0}^{+\infty}{b_n}$ due serie a termini non negativi e $b_n \neq 0$ definitivamente. Se $\displaystyle\lim_{n \rightarrow +\infty}{\frac{a_n}{b_n}} = 1$ ($L \neq 0$ finito), allora le due serie hanno lo stesso comportamento.

**Teorema** (criterio integrale) Sia $f:[0, +\infty[\rightarrow\mathbb{R}$ positiva decrescente e integrabile su ogni segmento $[0, M]\ \ \forall M > 0$. Allora

1. $\displaystyle\sum_{n = 0}^{+\infty}{f(n)}$ converge $\iff$ $f$ è integrabile in senso generalizzato

2. $\displaystyle\sum_{n = 0}^{+\infty}{f(n)}$ diverge $\iff$ $\displaystyle\int_{0}^{+\infty}{f(x)dx} = +\infty$

**Dimostrazione** (criterio del rapporto)

$$\frac{a_{n+1}}{a_n} \leq l + \epsilon = k\ \ \forall n \geq \nu(\epsilon)$$
    
allora per $n \geq \nu(\epsilon)$
    
$$a_n \leq ka_{n-1} \leq k^2a_{n-2} \leq ... \leq k^{n - \nu}a_{\nu} = \frac{a_{\nu}}{k^{\nu}}k^n$$
    
Se $l < 1$ allora $k < 1$ e per il criterio del confronto la serie converge, altrimenti se $l < 1$ allora $k > 1$ e per il criterio del confronto la serie diverge.

## Serie resto

**Definizione** Sia $\displaystyle\sum_{n = 0}^{+\infty}{a_n}$ una serie a termini positivi convergente. La serie $\displaystyle\sum_{n = k + 1}^{+\infty}{a_n} = a_{k+1} + a_{k+2} + ...$ è detta **serie resto di indice k**.

**Proposizione** Sia $R_n$ la somma della serie resto. Se $l$ è la somma della serie originaria e $S_n$ le sue somme parziali, allora

$$R_n = l - S_n$$

**Osservazione** Ovviamente $R_n \rightarrow 0$. Se riusciamo a stimare $R_n$ possiamo stimare l'errore approssimando $l$ con $S_n$.

## Altri criteri di convergenza

**Teorema** (criterio di Leibniz) (serie a segno alterno)
Sia $\{a_n\}_{n\in\mathbb{N}}$ una successione monotona decrescente e infinitesima. Allora la serie $\displaystyle\sum_{n = 0}^{+\infty}{(-1)^na_n}$ è convergente. Inoltre $|R_n| \leq |a_{n+1}|$.

**Dimostrazione** Siccome $\{a_n\} \searrow 0$, allora $a_{2n+1} \geq a_{2n+2}$, quindi

$$S_{2n+2} = S_{2n} - a_{2n+1} + a_{2n+2}$$

Ma

$$S_{2n} - a_{2n+1} + a_{2n+2} \leq S_{2n}$$

quindi la successione delle somme parziali di indici pari è decrescente.

Allo stesso modo

$$S_{2n+1} = S_{2n-1} + a_{2n} - a_{2n+1}$$

e

$$S_{2n-1} + a_{2n} - a_{2n+1} \geq S_{2n-1}$$

quindi la successione delle somme parziali di indici dispari è decrescente. Inoltre $\forall n \geq 1$

$$S_1 \leq S_3 \leq ... \leq S_{2n-1} \leq S_{2n} \leq S_{2n-2} \leq ... \leq S_0$$

Così $\{S_{2n}\}$ è convergente perché decrescente e inferiormente limitata; $\{S_{2n+1}\}$ è convergente perché crescente e superiormente limitata.

Ma $S_{2n} - S_{2n-1} = a_{2n} \rightarrow 0$ e quindi detti $L_1$ e $L_2$ i limiti rispettivamente di $S_{2n}$ e $S_{2n-1}$ si ha $L_1 = L_2 = l$. Quindi

$$S_n \rightarrow \displaystyle\sum_{n = 0}^{+\infty}{(-1)^na_n}$$

Si ha anche che:

- se $n$ è pari, $S_{n+1} \leq l \leq S_n$, quindi $|l - S_n| \leq |S_{n+1} - S_n| = |a_{n+1}|$

- se $n$ è dispari, $S_{n} \leq l \leq S_{n+1}$, quindi $|l - S_n| \leq |S_{n+1} - S_n| = |a_{n+1}|$

Quindi $|R_n| \leq |a_{n+1}|$.

**Teorema** (criterio di Dirichlet) Sia $\displaystyle\sum_{n = 0}^{+\infty}{z_n}$, $z_n\in\mathbb{C}$, una serie a somme parziali limitate ($|S_n| \leq M\ \ \forall n\in\mathbb{N}$). Se $\{b_n\} \searrow 0$ la serie $\displaystyle\sum_{n = 0}^{+\infty}{z_nb_n}$ converge.

**Teorema** (criterio di Abel) Sia $\displaystyle\sum_{n = 0}^{+\infty}{z_n}$, $z_n\in\mathbb{C}$ convergente e sia $\{b_n\}_{n\in\mathbb{N}}$ monotona limitata. Allora $\displaystyle\sum_{n = 0}^{+\infty}{z_nb_n}$ è convergente.

## Serie di funzioni

**Definizione** Siano $f_k: I \rightarrow \mathbb{R}$, $I\subseteq\mathbb{R}$ intervallo. La successione di funzioni $\{S_n(x)\}_{n\in\mathbb{N}}$, $S_n:I \rightarrow \mathbb{R}$ così definita

$$S_n(x) = \sum_{k = 0}^{n}{f_k(x)}$$

è detta **serie di funzioni** di termine generale $f_k(x)$.

**Definizione** Posto, se esiste finito, $\displaystyle S(x) = \lim_{n \rightarrow +\infty}{S_n(x)}$, si ha $S:A \subseteq I \rightarrow \mathbb{R}$ e si dice che $S(x)$ è la somma della serie di funzioni e si scrive:

$$S(x) = \sum_{k = 0}^{n}{f_k(x)}$$

Se la successione $\{S_n(x)\}_{n\in\mathbb{N}}$ converge puntualmente (risp. uniformemente) a $S(x)$ su $I$, allora si dice che la serie converge a $S(x)$ puntualmente (risp. uniformemente).

**Teorema** (criterio di Weierstrass) Sia $f_k: I \rightarrow \mathbb{R}$ una successione di funzioni, $I\subseteq\mathbb{R}$ e supponiamo che il $\sup_{i}{|f_k(x)|} = M_k < +\infty$ e $\sum_{k = 0}^{+\infty} < +\\infty$ (si dice che $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)}$ converge totalmente) allora $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)}$ converge assolutamente e uniformemente in $I$, ovvero converge uniformemente la serie di funzioni $\displaystyle\sum_{k = 0}^{+\infty}{|f_k(x)|}$.

### Proprietà delle serie di funzioni uniformemente convergenti

1. Se $f_k: I \rightarrow \mathbb{R}$ è una successione di funzioni continue in $x_0 \in I$ e $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)} \overset{\text{unif.}}{\rightarrow} S(x)$ su $I$, allora $S(x)$ è continua in $x_0$ e si scrive

    $$lim_{x \rightarrow x_0}{(\sum_{k = 0}^{+\infty}{f_k(x)})} = S(x_0) = \sum_{k = 0}^{+\infty}{\lim_{x \rightarrow x_0}{f_k(x)}}$$

2. Se $f_k:[a,b] \rightarrow \mathbb{R}$ è integrabile secondo Riemann e se $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)}$ converge uniformemente a $S(x)$ in $I$, allora $S$ è Riemann integrabile su $[a,b]$ e

    $$\int_{a}^{b}{S(x)dx} = \sum_{k = 0}^{+\infty}{\int_{a}^{b}{f_k(x)dx}} = \int_{a}^{b}{\sum_{k = 0}^{+\infty}{f_k(x)}}$$
    
3. Se $f_k \in \mathcal{C}^{(1)}(I, \mathbb{R})$, $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)}$ converge (per qualche $x_0 \in I$) e $\displaystyle\sum_{k = 0}^{+\infty}{f'_k(x)}$ converge uniformemente su $I$, allora $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)}$ converge uniformemente su $I$. Posto $\displaystyle S(x) = \sum_{k = 0}^{+\infty}{f_k(x)}$, $S(x)$ è derivabili e $\displaystyle S'(x) = \sum_{k = 0}^{+\infty}{f'_k(x)}$. Ovvero

    $$(\sum_{k = 0}^{+\infty}{f_k(x)})' = \sum_{k = 0}^{+\infty}{f'_k(x)}$$
    
## Serie di potenze

**Definizione** È detta serie di potenze di punto iniziale (o centro) $x_0\in\mathbb{R}$ la serie di funzioni $\displaystyle\sum_{k = 0}^{+\infty}{f_k(x)}$, $f_k(x) = a_k(x-x_0)^k$, $x\in\mathbb{R}$ e $\{a_k\}$ successione reale.

**Osservazione** $\displaystyle\sum_{k = 0}^{+\infty}{a_k(x-x_0)^k}$ converge per $x = x_0$ e detta $S(x)$ la somma (laddove è definita) si ha che $S(x_0) = 0$.

**Lemma** Supponiamo che $x_0 = 0$, $\displaystyle\sum_{k = 0}^{+\infty}{a_kx^k}$ e che in $x_1 \neq 0$ la serie $\displaystyle\sum_{k = 0}^{+\infty}{a_kx_1^k}$ converga. Allora la serie $\displaystyle\sum_{k = 0}^{+\infty}{a_kx^k}$ converge assolutamente $\forall x < |x_1|$.

**Dimostrazione** Poiché $\displaystyle\sum_{k = 0}^{+\infty}{a_kx_1^k}$ converge si ha $\displaystyle\lim_{n \rightarrow +\infty}{a_kx_1^k} = 0$, quindi $\exists M > 0$ tale che $|a_kx_1^k! \leq M\ \ \forall k \in \mathbb{N}$. Allora

$$|a_kx^k| = |a_kx_1^k(\frac{x}{x_1})^k| \leq M|(\frac{x}{x_1})^k|$$

ma $\frac{x}{x_1} < 1$ (x è fissato con |x| < |x_1|), perciò $\displaystyle\sum_{k = 0}^{+\infty}{|a_kx^k|}$ è maggiorata da una serie geometrica di ragione $< 1$ e quindi converge.

**Definizione** È detto **raggio di convergenza** della serie di potenze $\displaystyle\sum_{k = 0}^{+\infty}{a_kx^k}$ il numero:

$$R = \sup\{|x| : \sum_{k = 0}^{+\infty}{a_kx^k} \text{converge}\} \in [0, +\infty]$$

e l'intervallo $]-R, R[$ è detto intervallo di convergenza.

**Proposizione** Sia $\displaystyle\sum_{k = 0}^{+\infty}{a_kx^k}$ una serie di potenze di raggio di convergenza $R > 0$. Allora $\forall |x| < R$ la serie converge assolutamente.

**Dimostrazione** Segue dal lemma precedente.

**Proposizione** Sia $\displaystyle\sum_{k = 0}^{+\infty}{a_kx^k}$ una serie di potenze avente raggio di convergenza $R > 0$. Allora la serie converge assolutamente e uniformemente su ogni intervallo $[-r, r]$ con $r < R$.

**Dimostrazione** Per $x \in [-r, r]$ si ha $|a_kx^k| \leq |a_k|r^k$ e dal teorema di Weierstrass segue la tesi.

**Proposizione** Sia $R$ il raggio di convergenza di $\displaystyle\sum_{k = 0}^{+\infty}{a_kx^k}$, allora

1. se $\displaystyle\lambda = \lim_{n \rightarrow +\infty}{\sqrt[n]{|a_n|}} \implies R = \frac{1}{\lambda}$

2. se $\displaystyle\exists\lim_{n \rightarrow +\infty}{\left |\frac{a_{n+1}}{a_n} \right |} = \lambda \implies R = \frac{1}{\lambda}$

3. se $\displaystyle\exists\lim_{n \rightarrow +\infty}{\left |\frac{a_n}{a_{n+1}} \right |} = \lambda \implies R = \lambda$

**Teorema** Sia $f(x) = \sum_{k = 0}^{+\infty}{a_kx^k}$ con $|x| < R$, $R > 0$. Allora $f$ è derivabile e

$$f'(x) = \sum_{k = 1}^{+\infty}{ka_kx^{k-1}}$$

**Osservazione** Notiamo che la serie $\displaystyle\sum_{k = 1}^{+\infty}{ka_kx^{k-1}} = \sum_{k = 0}^{+\infty}{(k+1)a_{k+1}x^k}$ ha lo stesso raggio di convergenza.

**Osservazione** Possiamo iterare il teorema e ottenere

$$f^{(m)}(x) = \sum_{k = m}^{+\infty}{k(k-1)...(k-m+1)a_kx^{k-m}}$$

e $f\in\mathcal{C}^{(\infty)}(]-R, R[, \mathbb{R})$

**Proposizione** $\displaystyle\int_{a}^{b}{f(x)dx} = \sum_{k = 0}^{+\infty}{\int_{a}^{b}{a_kx^kdx}}$ con $[a,b] \subseteq ]-R, R[$.

**Osservazione** Se $R$ è raggio di convergenza non possiamo concludere nulla nei punti $x = \pm R$.

## Approssimazione di una funzione data mediante una serie opportuna di funzioni

### Serie di Taylor

**Definizione** (serie di Taylor) Siano $f\in\mathcal{C}^{(\infty)}(I, \mathbb{R})$ e $R$ raggio di convergenza di

$$\sum_{k = 0}^{+\infty}{\frac{f^{(k)}(x_0)}{k!}(x-x_0)}$$

Se vale

$$f(x) = \sum_{k = 0}^{+\infty}{\frac{f^{(k)}(x_0)}{k!}(x-x_0)},\ \ \ x_0 \in ]x_0-R, x_0+R[$$

allora tale serie si definisce **serie di Taylor** di punto iniziale $x_0$ sull'intervallo $]x_0-R,x_0+R[$ e diremo che $f$ è sviluppabile in serie di Taylor su tale intervallo. Inoltre si ha che, indicato con $T_n(x)$ il polinomio di Taylor di $f(x)$ di punto iniziale $x_0$, posto

$$R_n(x) = f(x) - T_n(x)$$

si ha

$$\lim_{n \rightarrow +\infty}{R_n(x)} = 0 \iff \left [f(x) = sum_{k = 0}^{+\infty}{\frac{f^{(k)}(x_0)}{k!}(x-x_0)} \right]$$

**Teorema** Sia $f\in\mathcal{C}^(1)(]x_0-\delta, x_0+\delta, \mathbb{R})$. Se $\exists M > 0$ tale che $|f^{(k)}(x_0)| < M^n\ \ \forall x \in ]x_0-\delta, x_0+\delta[$, allora $f$ è sviluppabile in serie di Taylor di punto iniziale $x_0$ in tutto l'intervallo.

**Dimostrazione** Dalla formula di Taylor con resto di Lagrange si ha

$$R_n(x) = \frac{f^{(n+1)}(c)}{(n+1)!}(x-x_0)^{n+1}$$

con $c$ compreso tra $x$ e $x_0$. Quindi

$$\forall x \in ]x_0-\delta, x_0+\delta[\ \ |R_n(x)| \leq \frac{M^{n+1}}{(n+1)!}(2\delta)^{n+1} = \frac{(2\delta M)^{n+1}}{(n+1)!} \rightarrow 0$$

Dalla definizione di serie di Taylor segue la tesi.

### Serie di Fourier

**Definizione** Sia $f:\mathbb{R}\rightarrow\mathbb{R}$, diremo che $f$ è periodica di periodo $T > 0$ se $f(x+T) = f(x)\ \ \forall x \in \mathbb{R}$. $T$ è detto periodo di $f$.

**Osservazione** Se $f$ è periodica di periodo $T$, allora è periodica di periodo $nT\ \ \forall n \in \mathbb{N}$. Se $S > 0$ è un altro periodo di $f$, allora $S+T$ è ancora un periodo di $f$. Se l'insieme dei periodi è discreto (non ha punti di accumulazione), su $[0, +\infty[$, è detto periodo fondamentale di $f$ il $\min\{S > 0 : S\ \text{è periodo di f}\}$.

**Osservazione** Le funzioni costanti sono periodiche di periodo $S \geq 0$ qualunque sia $S$.

**Osservazione** Le funzioni $\cos{(\frac{2\pi}{T}kx)}$ e $\sin{(\frac{2\pi}{T}kx)}$ sono periodiche di periodo $\frac{T}{k}$ e $T_f = T$.

**Osservazione** Posto $\omega = \frac{2\pi}{T}$ (frequenza) si ha

$$\int_{0}^{T}{\cos{(n\omega x)}\sin{(m\omega x)}dx} = 0\ \ \forall n. m \in \mathbb{N}$$

$$\int_{0}^{T}{\cos{(n\omega x)}\cos{(m\omega x)}dx} = \frac{T}{2}\delta{(n-m)}\ \ \forall n, m \in \mathbb{N}$$

$$\int_{0}^{T}{\sin{(n\omega x)}\sin{(m\omega x)}dx} = \frac{T}{2}\delta{(n-m)}\ \ \forall n, m \in \mathbb{N}$$

con $\delta = \left\{\begin{matrix}
1 & \text{se}\ x = 0\\ 
0 & \text{se}\ x = 1
\end{matrix}\right.$

**Dimostrazioni** Calcolare gli integrali ricordandosi che $\cos(x) = \frac{\exp{(ix)} + \exp{(-ix)}}{2}$ e $\sin(x) = \frac{\exp{(ix)} - \exp{(-ix)}}{2i}$.

**Definizione** (serie di Fourier) Sia $f:\mathbb{R}\rightarrow\mathbb{R}$ periodica di periodo fondamentale $T = T_f$ e supponiamo che in $[0, T]$ sia continua a tratti (ossia $f$ limitata e con un numero finito di punti di discontinuità di I specie). Definiamo le successioni $\{a_n\}_{n\in\mathbb{N}}$ e $\{b_n\}_{n\in\mathbb{N}}$ dei suoi coefficienti di Fourier in questo modo:

$$a_0 = \frac{2}{T}\int_{0}^{T}{f(x)dx}$$

$$a_n = \frac{2}{T}\int_{0}^{T}{f(x)\cos{(\frac{2\pi}{T}nx)}dx}$$

$$b_n = \frac{2}{T}\int_{0}^{T}{f(x)\sin{(\frac{2\pi}{T}nx)}dx}$$

La serie

$$\frac{a_0}{2} + \sum_{n = 1}^{+\infty}{\left [ a_n\cos{(\frac{2\pi}{T}nx)} + b_n\sin{(\frac{2\pi}{T}nx)} \right ]}$$

è detta **serie di Fourier** associata a $f$.

**Definizione** Sia $f:\mathbb{R}\rightarrow\mathbb{R}$ periodica di periodo fondamentale $T > 0$. Diremo che $f$ è $\mathcal{C}^{(1)}$ a tratti in $[0, T]$ se $f$ è derivabile in $[0, T]$ eccetto in un numero finito di punti più un numero finito di punti di discontinuità di I specie e nei punti di non derivabilità la sua derivata prima ammette limite destro e limite sinistro finiti.

**Teorema** (di Fourier) Sia $f:\mathbb{R}\rightarrow\mathbb{R}$ periodica di periodo fondamentale $T>0$ e $\mathcal{C}^{(1)}$ a tratti. Allora

- $f$ è sviluppabile in serie di Fourier in tutti i punti di continuità di $f$.

- se $c\in\mathbb{R}$ è un punto di discontinuità per $f$ allora la sua serie di Fourier converge a $\frac{1}{2}(f(c+) + f(c-))$

- $f$ è sviluppabile in serie di Fourier uniformemente su tutti gli intervalli di $\mathbb{R}$ che non contengono punti di discontinuità

- la serie di Fourier associata a $f$ è integrabile su ogni intervallo $[a,b]\subseteq\mathbb{R}$ e si ha

    $$\int_{a}^{b}{f(x)dx} = \frac{a_0}{2}(b-a) + \sum_{n = 1}^{+\infty}{\int_{a}^{b}{\left [ a_n\cos{(\frac{2\pi}{T}nx)} + b_n\sin{(\frac{2\pi}{T}nx)} \right ]}}$$
